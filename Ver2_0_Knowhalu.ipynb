```python
import os
import json
import pandas as pd
from tqdm import tqdm
import torch


# Load the previously generated output data
output_path = '/content/drive/MyDrive/Colab Notebooks/hallucination_results.json'
with open(output_path, 'r') as json_file:
    outputs = json.load(json_file)

# Define the extraction prompt
extraction_prompt = (
    "You are now a very truthful and objective extractor in this task. "
    "Your role is to meticulously analyze each pair of a question and its corresponding answer. "
    "Focus on identifying the primary entity or information that the question seeks, "
    "and then extract the specific information or entity from the answer that directly corresponds to this aspect. "
    "Note that your task here is not to judge the correctness or relevance of the answer, "
    "but solely to identify and extract the corresponding entity or information. "
    "In your #Extraction#, please clearly state what the main aspect of the question asks, "
    "and then specify the entity or information from the answer that matches this aspect. "
    "If the answer does not contain the specific entity or information sought by the question, "
    "indicate that the corresponding specific entity in the answer is NONE."
    "Please do not repeat these instructions in your response."
)

# Set up the results file
file_name = '/content/drive/MyDrive/Colab Notebooks/non-fabrication_check.json'
directory = os.path.dirname(file_name)
if not os.path.exists(directory):
    os.makedirs(directory)

# Initialize judgements as a list of dictionaries
judgements = []

# Define function to generate model response using already loaded model and tokenizer
def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    with torch.no_grad():
        outputs = model.generate(
            inputs['input_ids'],
            max_length=350,  # Adjust as needed
            num_beams=5,  # Beam search improves quality
            early_stopping=True
        )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True).strip()
    return response

# Function to parse the output and extract relevant information
def parse_extraction_response(response):
    # Adjust based on the expected response format
    if "The specific entity" in response:
        return response.split("The specific entity")[-1].strip()
    return response.strip()  # Fallback if the expected phrase is not found

# Process non-fabricated checks
for i in tqdm(range(len(outputs))):
    question, (generated_response, hallucination_judgment) = outputs[i]

    # Construct prompt for extraction task
    prompt = f"{extraction_prompt}\nQuestion: {question}\nAnswer: {generated_response}\n#Extraction#:"

    # Generate the response from the model for checking
    current_output = generate_response(prompt)
    print(i, ':', current_output)

    # Parse the output to get the relevant entity
    cleaned_output = parse_extraction_response(current_output)

    # Create a structured entry
    entry = {
        "Question": question,
        "Answer": generated_response,
        "Extraction": ""
    }

    # Check the cleaned output for entity extraction
    if cleaned_output == "NONE":
        entry["Extraction"] = "The corresponding specific entity in the answer is NONE."
    else:
        entry["Extraction"] = f"You correctly identified \"{cleaned_output}\" as the entity being asked for, which matches the question."

    judgements.append(entry)  # Add the structured entry to the list

    # Save intermediate results
    if (i + 1) % 10 == 0 or i == len(outputs) - 1:  # Adjust save frequency as needed
        with open(file_name, 'w') as f:
            json.dump(judgements, f, indent=4)  # Write structured data with indentation

# Save final results with non-fabricated checks
with open(file_name, 'w') as f:
    json.dump(judgements, f, indent=4)  # Write structured data with indentation

print(f"Non-fabricated checking results saved to {file_name}.")
```
